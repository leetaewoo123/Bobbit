```python
def ProductPrice(name):
    Dict = {}
    for year in range(18, 23):
        for month in range(1, 13):
            if month>9:
                DateCode = '20'+str(year)+str(month)+'4'
            else:
                DateCode = '20'+str(year)+'0'+str(month)+'4'

            CateCode = 'L002003'
            url = f"https://www.kpi.or.kr/www/life/info_detail.asp?CateCode={CateCode}&DateCode={DateCode}"
            response = requests.post(url)
            dom = BeautifulSoup(response.content.decode('euc-kr'), "html.parser")
            element_names = dom.select('.life-td-pa > td:nth-child(1)')
            for i in range(len(element_names)):
                if element_names[i].text == name:
                    element_prices = dom.select('.life-td-pa > td:nth-child(5)')
                    if month>9:
                        Dict['20'+str(year)+'년 '+str(month)+'월'] = element_prices[i].text
                    else:
                        Dict['20'+str(year)+'년 ' +str(month)+'월'] = element_prices[i].text
    df = pd.DataFrame({'날짜': list(Dict.keys()),
                    'price': list(Dict.values())})
    df.to_csv(name+'.csv')
```

```python
from selenium import webdriver
import time
from datetime import date
date.today().month
```

```python
driver = webdriver.Chrome(executable_path='/Users/joyeonglae/develop/chromedriver')
driver.get('http://price.consumer.or.kr/DR1001/DR1004_01.php')
date_Dict= {2:'2022', 3:'2021', 4:'2020', 5:'2019', 6:'2018', 7:'2017', 8:'2016', 9:'2015',
            10:'2014', 11:'2013', 12:'2012', 13:'2011'}
result = pd.DataFrame(columns=['날짜', 'price'])

for i in range(2, 14):

    time.sleep(0.5)
    driver.find_element_by_css_selector(f"#select01 > option:nth-child({i})").click()
    time.sleep(0.5)
    if i == 2:
        end = date.today().month
    else:
        end = 13
    driver.find_element_by_css_selector(f"#select02 > option:nth-child({end})").click()
    time.sleep(0.5)
    driver.find_element_by_css_selector("#select03 > option[value='Item0202']").click()
    time.sleep(0.5)
    driver.find_element_by_css_selector("#body > div.section > div.box02 > ul > li.btn01 > a").click()
    time.sleep(1.0)
    elements = driver.find_elements_by_css_selector("#body > div.section > div.item > table > tbody > tr")
    datas =[]
    cnt = 0
    for element in elements:
        if element.find_element_by_css_selector('td:nth-child(2)').text == '[농심]신라면' and \
            '5개' in element.find_element_by_css_selector('td:nth-child(3)').text:
            
            for j in range(end-1, 0, -1):
                if j>9:
                    data = {
                        '날짜': date_Dict[i] + '년 ' + str(j) + '월',
                        'price': element.find_element_by_css_selector(f"td:nth-child({j+3})").text,
                    }
                    datas.append(data)
                else:
                    data = {
                        '날짜': date_Dict[i] + '년 ' + str(j) + '월',
                        'price': element.find_element_by_css_selector(f"td:nth-child({j+3})").text,
                    }
                    datas.append(data)
            

    df = pd.DataFrame(datas)
    result= pd.concat([result, df])
```

```python
result.reset_index(inplace=True)
```

```python
result = result.drop('index', axis=1)
```

```python
result.to_csv('신라면2.csv')
```
